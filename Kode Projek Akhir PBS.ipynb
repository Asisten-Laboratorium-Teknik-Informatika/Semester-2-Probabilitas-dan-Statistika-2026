```python
# Pertemuan 4: Deskriptif Statistik (Ukuran Pemusatan Data)

# 1. Track Popularity
mean_track = df['track_popularity'].mean()
median_track = df['track_popularity'].median()
modus_track = df['track_popularity'].mode()[0]

print("Ukuran Pemusatan Data pada Track Popularity")
print(f"Mean  : {mean_track:.2f}")
print(f"Median: {median_track}")
print(f"Modus : {modus_track}")

# 2. Artist Popularity
mean_artist = df['artist_popularity'].mean()
median_artist = df['artist_popularity'].median()
modus_artist = df['artist_popularity'].mode()[0]

print("\nUkuran Pemusatan Data pada Artist Popularity")
print(f"Mean  : {mean_artist:.2f}")
print(f"Median: {median_artist}")
print(f"Modus : {modus_artist}")

# 3. Album Total Tracks
mean_album = df['album_total_tracks'].mean()
median_album = df['album_total_tracks'].median()
modus_album = df['album_total_tracks'].mode()[0]

print("\nUkuran Pemusatan Data pada Track Popularity")
print(f"Mean  : {mean_album:.2f}")
print(f"Median: {median_album}")
print(f"Modus : {modus_album}")


# Pertemuan 4: Deskriptif Statistik (Ukuran Penyebaran Data)

# 1. Track Popularity
jangkauan_track = df['track_popularity'].max() - df['track_popularity'].min()
simpangan_rata_rata_track = np.mean(np.abs(df['track_popularity'] - df['track_popularity'].mean()))
varians_track = df['track_popularity'].var()
std_deviasi_track = df['track_popularity'].std()

print("\nUkuran Penyebaran Data pada Track Popularity")
print(f"Jangkauan (Range)   : {jangkauan_track}")
print(f"Simpangan Rata-Rata : {simpangan_rata_rata_track:.2f}")
print(f"Varians (Ragam)     : {varians_track:.2f}")
print(f"Standar Deviasi     : {std_deviasi_track:.2f}")

# 2. Artist Popularity
jangkauan_artist = df['artist_popularity'].max() - df['artist_popularity'].min()
simpangan_rata_rata_artist = np.mean(np.abs(df['artist_popularity'] - df['artist_popularity'].mean()))
varians_artist = df['artist_popularity'].var()
std_deviasi_artist = df['artist_popularity'].std()

print("\nUkuran Penyebaran Data pada Artist Popularity")
print(f"Jangkauan (Range)   : {jangkauan_artist}")
print(f"Simpangan Rata-Rata : {simpangan_rata_rata_artist:.2f}")
print(f"Varians (Ragam)     : {varians_artist:.2f}")
print(f"Standar Deviasi     : {std_deviasi_artist:.2f}")

# 3. Album Total Tracks
jangkauan_album = df['album_total_tracks'].max() - df['album_total_tracks'].min()
simpangan_rata_rata_album = np.mean(np.abs(df['album_total_tracks'] - df['album_total_tracks'].mean()))
varians_album = df['album_total_tracks'].var()
std_deviasi_album = df['album_total_tracks'].std()

print("\nUkuran Penyebaran Data pada Album Total Tracks")
print(f"Jangkauan (Range)   : {jangkauan_album}")
print(f"Simpangan Rata-Rata : {simpangan_rata_rata_album:.2f}")
print(f"Varians (Ragam)     : {varians_album:.2f}")
print(f"Standar Deviasi     : {std_deviasi_album:.2f}")

# Pertemuan 4: Deskriptif Statistik (Bentuk Distribusi)

# 1. Track Popularity
skew_track = df['track_popularity'].skew()
kurt_track = df['track_popularity'].kurt()

print("\nBentuk Distribusi Data pada Track Popularity")
print(f"Skewness (Kemiringan)  : {skew_track:.2f}")
print(f"Kurtosis (Keruncingan) : {kurt_track:.2f}")

# 2. Artist Popularity
skew_artist = df['artist_popularity'].skew()
kurt_artist = df['artist_popularity'].kurt()

print("\nBentuk Distribusi Data pada Artist Popularity")
print(f"Skewness (Kemiringan)  : {skew_artist:.2f}")
print(f"Kurtosis (Keruncingan) : {kurt_artist:.2f}")

# 3. Album Total Tracks
skew_album = df['album_total_tracks'].skew()
kurt_album = df['album_total_tracks'].kurt()

print("\nBentuk Distribusi Data pada Album Total Tracks")
print(f"Skewness (Kemiringan)  : {skew_album:.2f}")
print(f"Kurtosis (Keruncingan) : {kurt_album:.2f}")
```

```python
# Pertemuan 3: Simulasi Monte Carlo
# Monte Carlo Track Popularity
data_track = df['track_popularity'].dropna()

n_simulasi = 10000
hasil_simulasi = []

for i in range(n_simulasi):
    sample = np.random.choice(data_track, size = 50, replace = True)
    rata2 = np.mean(sample)
    hasil_simulasi.append(rata2)

hasil_simulasi = np.array(hasil_simulasi)

print("Rata-rata simulasi:", np.mean(hasil_simulasi))
print("Standar deviasi:", np.std(hasil_simulasi))

plt.hist(hasil_simulasi, bins = 50, density = True, color = 'yellow', edgecolor = 'black')
plt.title("Simulasi Monte Carlo")
plt.xlabel("Rata-rata Track Popularity")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Monte Carlo Artist Popularity
data_artist = df['artist_popularity'].dropna()

n_simulasi = 10000
hasil_simulasi = []

for i in range(n_simulasi):
    sample = np.random.choice(data_artist, size = 50, replace = True)
    rata2 = np.mean(sample)
    hasil_simulasi.append(rata2)

hasil_simulasi = np.array(hasil_simulasi)

print("Rata-rata simulasi:", np.mean(hasil_simulasi))
print("Standar deviasi:", np.std(hasil_simulasi))

plt.hist(hasil_simulasi, bins = 50, edgecolor = 'black')
plt.title("Simulasi Monte Carlo")
plt.xlabel("Rata-rata Artist Popularity")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Monte Carlo Album Total Tracks
data_album = df['album_total_tracks'].dropna()

n_simulasi = 10000
hasil_simulasi = []

for i in range(n_simulasi):
    sample = np.random.choice(data_album, size = 50, replace = True)
    rata2 = np.mean(sample)
    hasil_simulasi.append(rata2)

hasil_simulasi = np.array(hasil_simulasi)

print("Rata-rata simulasi:", np.mean(hasil_simulasi))
print("Standar deviasi:", np.std(hasil_simulasi))

plt.hist(hasil_simulasi, bins = 50, density = True, color = 'red', edgecolor = 'black')
plt.title("Simulasi Monte Carlo")
plt.xlabel("Rata-rata Album Total Tracks")
plt.ylabel("Frekuensi")
plt.show()
```

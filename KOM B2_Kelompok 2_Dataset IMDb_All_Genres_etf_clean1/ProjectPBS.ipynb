```python
plt.figure(figsize=(8,5))
plt.hist(df['Rating'], bins=20)
plt.title('Distribusi Rating Film')
plt.xlabel('Rating')
plt.ylabel('Jumlah Film')

plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.boxplot(df['Runtime(Mins)'])
plt.title('Boxplot Runtime Film')

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Runtime(Mins)'], df['Rating'])
plt.title('Runtime vs Rating')
plt.xlabel('Runtime (Minutes)')
plt.ylabel('Rating')

plt.show()
```

```python
top_genre = df['main_genre'].value_counts().head(10)
plt.figure(figsize=(10,5))
top_genre.plot(kind='bar')
plt.title('Top 10 Genre Film')
plt.xlabel('Genre')
plt.ylabel('Jumlah Film')

plt.show()
```
